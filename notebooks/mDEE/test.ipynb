# 2D Common Corruptions

```python
# !pip3 install imagecorruptions
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from imagecorruptions import corrupt

path = "1.jpg"

image = Image.open(path)
image_array = np.array(image)
plt.imshow(image)
```

```python
image_array.shape
```

```python

corrupted_image = corrupt(image_array, corruption_name='gaussian_blur', severity=2)
plt.imshow(corrupted_image)
```

```python
from imagecorruptions import get_corruption_names


for corruption in get_corruption_names():
    for severity in range(5):
        corrupted = corrupt(image_array, corruption_name=corruption, severity=severity+1)
        plt.imshow(corrupted);
```
